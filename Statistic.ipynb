```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('data/im.db')
cur = conn.cursor()
```

```python
pd.read_sql("""
SELECT * 
  FROM movie_basics;
""", conn).head(20)
```

# Average minutes over 700000 voters

```python
pd.read_sql("""
SELECT AVG(runtime_minutes) 
  FROM movie_basics;
""", conn).head(20)
```

```python
pd.read_sql("""
SELECT *
FROM movie_ratings
ORDER BY numvotes DESC;
""", conn).head()
```

# Average rating votes over 700000 voters

```python
pd.read_sql("""
SELECT AVG(averagerating)
FROM movie_ratings
WHERE numvotes > 700000;
""", conn).head(20)
```
